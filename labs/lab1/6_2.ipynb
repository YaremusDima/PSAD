Недостатки подхода правительства:

- Было бы лучше, если бы выборки были связаны, то есть каждая из трех марок автомобиля давалась бы одному и тому же человеку. Также нужно контнолировать, что результаты должны быть проведены в одном  и том же городе и в один и тот же месяц
- Размер выборки в 20 человек может быть маловат при использовании ассимптотических критериев

Вариант более корректного решения задачи: 

Исходя из недостатков, нужно увеличить объем выборки, скажем, до 100 человек, при этом каждому из них будет дана возможность поездить на всех  трех марках автомомбилей. Далее замерить результаты в течение одного и того же месяца в одном и том же городе.

Формально:

Проверяем попарные гипотезы

$H_0$: средние скорости передвижения для каждой пары выборок равны.

$H_1$: средние скорости передвижения для каждой пары выборок различаются.

уровень значимости $\alpha = 0.05$
Для учета множественности гипотез воспользуемся поправкой Бонферрони, при которой уровень значимости для каждой гипотезы будет равен α / k, где k - количество сравнений.

Для 3 пар выборок будет 3 сравнения, поэтому уровень значимости для каждой гипотезы будет α / 3 = 0.0167.
