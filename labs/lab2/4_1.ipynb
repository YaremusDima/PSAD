## Задача 4.1
На основе представленной ниже языковой модели требуется:
1. Выбрать наиболее вероятную последовательность слов, которая начинается со слова `<start>` и заканчивается словом `<end>`.
2. Выбрать последовательность токенов, которая начинается со слова `<start>` и заканчивается словом `<end>` и имеет максимальный $score = \frac{1}{l}\sum_{i=1}^{l} \log p(w_i|w_{i-1})$, где `l` задает длину полученной последовательности.
```
p(обучение|<start>)=0.14
p(<end>|машинный)=0.5
p(<end>|обучение)=0.5
p(не|обучение)=0.5
p(без|возможно)=0.5
p(без|<start>)=0.14
p(не|<start>)=0.14
p(<end>|без)=0.5
p(<end>|не)=0.5
p(обучение|машинный)=0.5
p(<end>|статистика)=1.0
p(возможно|не)=0.5
p(возможно|<start>)=0.14
p(машинный|<start>)=0.29
p(статистика|<start>)=0.14
p(статистика|без)=0.5
p(<end>|возможно)=0.5
```
Все выкладки должны быть сделаны аналитически, без использования компьютера.
*Важно:* решением при помощи полного перебора не будет оцениваться в полный балл.

### 4.1.1

Заметим, что сумма переходов из `<start>` равна 0.99, скорее всего, это из-за того, что веса были 1/7, 1/7, ..., 2/7. Поэтому 0.28571428571 округлилось в 0.29, а 0.14285714285 в 0.14, так что всё окей. Циклов также нет.

Очевидно, общая вероятность посделовательности это произведение вероятностей переходов, первый из который идет из `<start>`, а последний входит в `<end>`. 

Исходя из этого, можем утверждать, что верятнотсть последовательности `<start> -> статистика` <= 0.14. Аналогично для других слов из `<start>`, кроме `<start> -> машинный` у которого вероятность <=
0.29. Далее может быть либо `<start> -> машинный -> <end>`, либо `<start> -> машинный -> обучение`, у первой вероятность == 0.14, у второй <=0.14. Таким образом, у всех возможнных последовательностей вероятность меньше или равна 0.14, так что если найдем последовательность с вероятностью 0.14, это и будет одна из самых вероятных последовательностей, то есть `<start> -> машинный -> <end>`

Ответ: `<start> -> машинный -> <end>`

### 4.1.2

Для начала заметим, что все вероятности переходов, которые не включают слова `<start>` и `<end>` имеют вероятность 0.5. Также заметим, что вроятность перейти в `<end>` у всех слов, кроме `статистика`(у неё 1.0), равна 0.5

Преобразуем score

$$
score = \frac{1}{l}\sum_{i=1}^{l} \log p(w_i|w_{i-1}) = 
\log \sqrt[l]{\prod_{i=1}^{l}p(w_i|w_{i-1})} = 
\log \sqrt[l]{p(w_1|\text{<start>}) \cdot p(\text{<end>}|w_{l-1}) \cdot\prod_{i=2}^{l-1} p(w_i|w_{i-1})} = \\ =
\log \sqrt[l]{p(w_1|\text{<start>}) \cdot p(\text{<end>}|w_{l-1}) \cdot 0.5^{l-2}} = 
\log 0.5\sqrt[l]{p(w_1|\text{<start>}) \cdot p(\text{<end>}|w_{l-1}) \cdot 4} 
$$

Будем максимизировать $score^* = \sqrt[l]{4 \cdot p(w_1|\text{<start>}) \cdot p(\text{<end>}|w_{l-1}) } $

Получается, наш скор зависит от первого слова, последнего слова и количества слов в предложении.

Переберем все варианты:

- $w_1$ = машинный, $w_{l-1}$ = статистика: $score = \arg\min\limits_{l\in\{8\}} \sqrt[l]{\frac{4 \cdot 2 \cdot 1}{7}} \approx 1.017 $

- $w_1$ = машинный, $w_{l-1}$ != статистика: $score^* = \arg\min\limits_{l\in\{7,6,5,4,3\}} \sqrt[l]{\frac{4 \cdot 1 \cdot 1}{7}} \approx 0.830 $ при $l = 3$

- $w_1$ != машинный, $w_{l-1}$ = статистика: $score^* = \arg\min\limits_{l\in\{7,6,5,4,3\}} \sqrt[l]{\frac{4 \cdot 2 \cdot 1}{7 \cdot 2}} \approx 0.830 $ при $l = 3$

- $w_1$ != машинный, $w_{l-1}$ != статистика:  $score^* = \arg\min\limits_{l\in\{6,5,4,3\}} \sqrt[l]{\frac{4 \cdot 1 \cdot 1}{7 \cdot 2}} \approx 0.659 $ при $l = 3$

Отсюда, самый высокий скор на предложении `<start> Машинный обучение не возможно без статистика <end>`
